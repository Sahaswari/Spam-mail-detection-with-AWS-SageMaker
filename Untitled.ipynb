```python
import pandas as pd
data = pd.read_csv("spambase.csv")
print(data.head())
```

```python
x = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
 import sagemaker
from sagemaker import LinearLearner

sagemaker_session = sagemaker.Session()
role = sagemaker.get_execution_role()

linear = LinearLearner(role=role, instance_count=1, instance_type='ml.m5.xlarge',predictor_type='binary_classifier')
```

```python
x_train_float32 = x_train.astype('float32')
x_test_float32 = x_test.astype('float32')
y_train_float32 = x_train.astype('float32')
y_test_float32 = x_test.astype('float32')
```

```python
train_records = linear.record_set(x_train_float32.values, labels=y_train_float32.values)

test_records = linear.record_set(x_test_float32.values, labels=y_test_float32.values)
```

```python
linear.fit(train_records)
```

```python
predictor = linear.deploy(initial_instance_count=1, instance_type='ml.m5.xlarge')
```
